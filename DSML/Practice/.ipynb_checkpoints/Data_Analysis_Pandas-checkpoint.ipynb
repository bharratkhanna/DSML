#### Task #1: Merging 12 months of data

```python
files = [file for file in os.listdir("Sales_Data/")]
for file in files: print(file)
```

```python
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data,df],axis = 0)

all_months_data.head()
```

#### Question #1: What was the best month of sales? How much was earned that month?

```python
results = sales_data.groupby("Months").sum()
results
```

```python
import matplotlib.pyplot as plt
```

```python
months = results.index.values
sales = results["Sales"]
plt.bar(months,sales)

plt.ticklabel_format(style='plain')
plt.xticks(months)
plt.ylabel("Sales in USD")
plt.xlabel("Months")
plt.title("Sales")

plt.show()
```

#####  Question #2: What US city has highest number of sale?

```python
backup_sales_data = sales_data.copy()
```

```python
sales_data.head()
```

#### Add city column

```python
def getcity(address):
    return address.split(',')[1]

def getstate(address):
    return address.split(',')[2].split(" ")[1]
    
sales_data["City"] = sales_data["Purchase Address"].apply(lambda x: getcity(x) + ' ' + getstate(x))
sales_data.head()
```

```python
backup_sales_data = sales_data.copy()
```

```python
results = sales_data.groupby("City").sum()
```

```python
results
```

```python
cities = results.index.values
sales = results["Sales"]
plt.bar(cities,sales)

plt.ticklabel_format(axis = "y", style="plain")
plt.ylabel("Sales")
plt.xlabel("Cities")
plt.xticks(rotation = "vertical")
plt.title("Sales by City")

plt.show()
```

#### Question #3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
sales_data.head()
```

```python
sales_data["Order Date"] = pd.to_datetime(sales_data["Order Date"])
```

```python
sales_data.head()
```

```python
sales_data["Hour"] = sales_data["Order Date"].dt.hour
```

```python
sales_data.head()
```

```python
results =  sales_data.groupby("Hour").count()
```

```python
results
```

```python
x = results.index.values
y = results["Sales"]
plt.plot(x,y)

plt.grid()
plt.title("Hourly Sales")
plt.xlabel("Time")
plt.ylabel("No of sales")
plt.xticks(x)
plt.show()
```

#### Question #4: What are the products most sold together?

```python
df = sales_data[sales_data["Order ID"].duplicated(keep =False)]
```

```python
df
```

```python
df["Grouped"]= sales_data.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))
df.head()
```

```python
df = df[["Order ID","Grouped"]].drop_duplicates()
```

```python
df4 = df[df["Grouped"] == "iPhone,Lightning Charging Cable"]
```

```python
df4.count()
```

```python
df4
```

```python
df2 = df.groupby("Grouped").count()
```

```python
df2.sort_values("Order ID", ascending = False)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(10)
```

#### What product sold the most? Why do you think so?

```python
backup_sales_data = sales_data.copy()
```

```python
most_sold_product =  sales_data.groupby("Product").sum()
```

```python
most_sold_product
```

```python
x = most_sold_product.index.values
y = most_sold_product["Quantity Ordered"]
plt.bar(x,y)

plt.title("Highest Sales")
plt.xlabel("Products")
plt.ylabel("Qunatity")
plt.xticks(rotation = "vertical")
plt.show()
```

```python
prices = sales_data.groupby("Product").mean()
prices["Price Each"]
```

```python
x = most_sold_product.index.values
y = most_sold_product["Quantity Ordered"]
x1 = x
y1 = prices["Price Each"]
plt.bar(x,y)
plt.bar(x1,y1)

plt.title("Highest Sales")
plt.xlabel("Products")
plt.ylabel("Qunatity")
plt.xticks(rotation = "vertical")
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x,y)
ax2.plot(x,y1, "b-", color ="orange" )

plt.title("Highest Sales/Price")
ax1.set_xlabel("Product Name")
ax1.set_ylabel("Quantity Ordered")
ax2.set_ylabel("Price ($)")
ax1.set_xticklabels(x ,rotation = "vertical")
plt.show()
```
