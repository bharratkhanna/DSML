# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris
```

### Distplot

```python
sns.distplot(iris['petal_length'])
```

### Histplot

```python
sns.histplot(iris["sepal_length"])
```

### Scatterplot

```python
# Scatter plot - when two continuous columns/ numerical columns
sns.scatterplot([iris["sepal_width"],iris["petal_width"]])
```

```python
# Lineplot- when one axis is time
sns.lineplot([iris.index,iris["sepal_length"]])
```

### Barplot

```python
# Barplot - when categorical values
sns.countplot(x=iris["species"])
```

### Boxplot

#### Outliers

```python
# Boxplot- Outliers
sns.boxplot([iris["petal_width"],iris["sepal_width"],iris["petal_length"],iris["sepal_length"]])
```

```python
sns.boxplot(y = iris["sepal_width"])
```

```python
sns.boxplot(x=iris["species"], y=iris["sepal_width"])
```

### Pairplot

```python
# Pairplot
sns.pairplot(iris, hue="species")
```

### Correlation

```python
corrDf = iris.corr()
corrDf.head()
```

### Heatmap

```python
sns.heatmap(iris.corr(),annot=True)
```
