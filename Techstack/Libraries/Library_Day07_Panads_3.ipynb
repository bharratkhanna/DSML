## DataFrame

```python
import pandas as pd
import numpy as np
```

```python
myDic = {'Emp_Id' : [101,101,103,104,105], 
         "Emp_Name" : ["Tej", "Palak", "Uzman", "Aayushi", "Sacin"], 
         "Gender" : ["M","F","M","F","M"], 
         "Salary" : [45000,35000,65000,55000,31000]}
```

```python
myDic["Emp_Id","Gender"]
```

```python
# Dataframe
# pd.DataFrame() -  we can create data frame using this method

myDf = pd.DataFrame(myDic)
myDf
```

```python
myList = [[111,"Jyoti", "DSA"],
          [112, "Rashi", "BA",],
          [113,"Bharat", "AI",],
          [114, "Mayaram", "DSA"]]

myList
```

```python
df2 = pd.DataFrame(myList, columns= ['Std_Roll',"Name","Course"])
df2
```

```python
# Series to Dataframe

ser1 = pd.Series(np.random.randint(20,50, 5))
ser2 = pd.Series(np.random.randint(35000,55000 , 5))

ser1
```

```python
df3 = pd.DataFrame({"Age" : ser1, "Salary" : ser2})
df3
```
