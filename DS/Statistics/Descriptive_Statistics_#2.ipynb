 ## Visualizing 

### Categorical

```python
import math as m
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
frequency = [124, 98, 113]
car_brands = ['Audi', 'BMW', 'Mercedes']
distribution = {
    "Car Brands": car_brands,
    "Frequency": frequency
}

freq_dist =  pd.DataFrame(distribution)
freq_dist
```

#### Bar Chart

```python
sns.barplot(x=freq_dist["Car Brands"], y = freq_dist["Frequency"], data = freq_dist)
```

#### Pie-chart

```python
palette_color = sns.color_palette('tab10')

plt.pie(freq_dist["Frequency"], labels=freq_dist["Car Brands"], colors=palette_color, autopct='%.0f%%')
plt.show()
```

#### Pareto Diagram 

```python
from matplotlib.ticker import PercentFormatter
df = freq_dist.copy()
df.sort_values("Frequency",ascending=False)
df["Cum Percentage"] = round(df["Frequency"].cumsum()/df["Frequency"].sum()*100,2)

fig, ax = plt.subplots()

# Plot bars (i.e. frequencies)
ax.bar(df["Car Brands"], df["Frequency"],color=palette_color)
ax.set_title("Pareto Chart")
ax.set_xlabel("Car_Brands")
ax.set_ylabel("Frequency");

# Second y axis (i.e. cumulative percentage)
ax2 = ax.twinx()
ax2.plot(df["Car Brands"], df["Cum Percentage"], color="orange", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylabel("Cumulative Percentage");
```

### Numerical

```python
numbers = np.random.randint(1,100,size=20)
freq_num = [1 for i in range(0,20)]

num_dist = pd.DataFrame({
    "Dataset": numbers,
    "Frequency": freq_num
})
num_dist.sort_values("Dataset",inplace=True,ignore_index=True)
num_dist
```

 **_Divide data in equal intervals_**:- 

### $$ \frac{\text{largest number - smallest number}}{\text{number of desired intervals}}$$

```python
interval_length = m.ceil((100-1)/5)
interval_length
```

```python
interval_start = [1,21,41,61,81]
interval_end = [21,41,61,81,101]
freq_num = [4,3,3,6,4]

num_dist = pd.DataFrame({
    "Interval Start": interval_start,
    "Interval End": interval_end,
    "Frequency": freq_num
})
num_dist
```

**_Relative Frequency_**:-

### $$\frac{\text{Frequency}}{\text{Total Frequency}}$$

```python
num_dist["Reltive Frequency"] = num_dist["Frequency"]/num_dist["Frequency"].sum()
num_dist
```

### Comparison between two variables/ continuous columns

#### Cross Tables

```python
types = ["Stocks", "Bonds", "Real Estate"]
inv_A = [96,181,88]
inv_B = [185,3,152]
inv_C = [39,29,142]

cross_table = pd.DataFrame({
    "Type of Investment\Investor": types,
    "Inverstor A": inv_A,
    "Investor B": inv_B,
    "Investor C": inv_C
})

cross_table["Total"] =  cross_table["Inverstor A"] + cross_table["Investor B"] + cross_table["Investor C"]
cross_table
```
