```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

#### Ignoring warnings

```python
warnings.filterwarnings('ignore')
```

#### Loading the file

```python
data = pd.read_csv("Titanic.csv")
data.head()
```

#### Explore the dataset

```python
print("#"*80)
print(data.shape)
print("#"*80)
print(data.info())
print("#"*80)
```

##### Outliers
- They can be judget by checking Interquartile range and max/min values

```python
print("#"*80)
print(data.describe().T)
print("#"*80)
```

#### Bivariate Analysis

```python
plt.figure()
sns.heatmap(data.corr(),vmin=-1,vmax=1,annot=True)
plt.show()
```

```python
plt.figure()
sns.pairplot(data=data,x_vars=["Fare","Sex","Pclass"], y_vars="Survived",kind="hist")
```

```python
sns.countplot(hue="Sex",x="Survived",data=data)
```
