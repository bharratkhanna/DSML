### Model Perfomance Methods
- Comparison can be done only by comparing two models
- For eg, one model dont drop VIF, in another drop VIF

#### Maximum Log Likelihood
- Gives perfomance of model
- But is boundless
- Result always in negative values
- Denoted by **LL**
- Higher the LL better the model
- Lower the LL worst the model

#### Log Loss
- Similar to Maximum Log Likelihood
- Values are in positive term
- Lower the -LL better the model
- Higher the -LL worst the model

$$ -1 \times (LL) $$

#### Deviance
- Lower the Deviance better the model
- Higher the Deviance worst the model

$$ D = -2 \times LL $$

#### AIC Akaike Information Criteria
-  Compares no of features in model too
- Lower the AIC better the model
- Higher the AIC worst the model

$$ AIC = -2LL + 2k $$
- k = Total no of independent variables

#### Base Information Criteria
-  Compares no of features & rows in model too
- Lower the BIC better the model
- Higher the BIC worst the model

$$ AIC = -2LL + 2k \times \log n $$
- k = Total no of independent variables
- n = Total no of rows

#### Null Deviance
- Not a single element is considered

#### Residual Deviance
- All variable are considdered

* if Null Deviance < Residual Deviance, **Discard the model**

### Assumptions of Logistic Regression
- Depedent variables should be categorical
- Presence of one or more Independent Variable
- There should no MultiCollinearity Relationship between Independent Variables

#### Accuracy

$$ \frac{\text{True Prediction}}{\text{Total Values}} $$
#### $$ Accuracy = \frac{\text{TP + TN}}{\text{TP + FN + FP + TN}} $$

- Value given in Percentage (Higher means better)
- Chances are that True Positive is identified by model but True Negative is not or vice-vers
- Hence we check Precision & Recall

#### Precision
$$ \frac{\text{True Positive}}{\text{Total Predicted Positive}} $$
#### $$ Precision  = \frac{\text{TP}}{\text{FP + TP}} $$

- Value given in Percentage (Higher means better)
- Lower FP means Higher Precison

#### Recall
$$ \frac{\text{True Positive}}{\text{Total Actual Positive}} $$
#### $$ Recall  = \frac{\text{TP}}{\text{FN + TP}} $$

- Value given in Percentage (Higher means better)
- Lower FN means Higher Recall

#### F1-Scope
- Harmonic mean of precision & recall
#### $$ \text{F1 Score} = \frac{-2 \times Precison \times Recall}{Precision + Recall} $$
