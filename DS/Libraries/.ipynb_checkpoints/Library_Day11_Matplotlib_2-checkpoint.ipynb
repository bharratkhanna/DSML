```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10)
plt.plot(x, [y**2 for y in x])
plt.show()
```

```python
# Can take n values of x,y for multi line plot
plt.plot(x,[x**2 for x in x], x,[x**3 for x in x], x,[x**4 for x in x]) 
plt.show()
```

### Grid

```python
# grid() - by default value is True
x = np.arange(5,6,0.1)
print(x)
plt.plot(x)
plt.grid(True)
```

### Axis

```python
# xlabel(), ylabel() 
plt.plot(x, [y**2 for y in x])
plt.grid(True)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

### Title

```python
# title()
plt.plot(x, [y**3 for y in x])
plt.grid(True)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("First-Graph")
```

### Axis

```python
# axis()- #First value (minx, miny),Second value (maxx,maxy)
plt.plot(x, [y*25 for y in x])
plt.title("Graph")
plt.axis((4.5,6.5,120,150)) 
```

```python
x = np.arange(1,10)
plt.plot(x,[x**2 for x in x], label = r'$y = x^2$')
plt.plot(x,[x**3 for x in x], label = r'$y = x^3$')
plt.plot(x,[x**4 for x in x], label = r'$y = x^4$') 
plt.title("Multi-Line")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```
