```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv("Data_File.csv")
df
```

### Na in Data Frame

```python
# find null values in df
df.isna().any()
```

```python
# Selecting column and implementing null negation
df["Variable_A"].fillna(df["Variable_A"].mean())

# Good Code
mean = df["Variable_A"].mean()
df["Variable_A"] = df["Variable_A"].fillna(mean)
```

```python
df.isna().sum()
```

```python
# Subsetting/ Extracting
# rows

df[2:7]
```

```python
# Column
df["Variable_A"]
```

```python
# Multiple Columns 
df[["Variable_A","Variable_B"]]
```

### Value Based Subsetting

```python
# Value Based Subsetting (R,C format)
# loc:-

df.loc[[1,3],["Variable_A","Variable_C"]]
```

### Conditions in Dataframe

```python
# Boolean that satisfies condition
df["Variable_A"] > 10 
```

```python
# Displays all value where Variable_A is greater than 5

df[df["Variable_A"] > 5]
```

```python
df[(df["A"] > 5 and df["B"] > 150)]
```

```python
# To show only one column with this condition
df["Variable_A"][df["Variable_A"] > 5]
```
