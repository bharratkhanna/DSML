```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x = np.random.randint(0,10,10)
x = np.arange(1,10)
x
```

```python
# y = np.random.randint(0,10,10)
y = np.arange(1,10)
y
```

### Single Line Graph

```python
# Line graph

plt.plot(x)
```

```python
# X^2 Graph
x = np.arange(1,10)
y = [i**2 for i in x]

plt.plot(x,y)
plt.show()
```

### Multi-Line Graph

```python
x = np.arange(1,10)

plt.plot(x,[x**2 for x in x])
plt.plot(x,[x**3 for x in x])
plt.plot(x,[x**4 for x in x])
plt.show()
```
