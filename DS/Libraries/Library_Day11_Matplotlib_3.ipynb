```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.0,6.0,0.01)
plt.plot(x,[x**2 for x in x], label = r'$y = x^2$')
plt.plot(x,[x**3 for x in x], label = r'$y = x^3$')
plt.plot(x,[x**4 for x in x], label = r'$y = x^4$') 

plt.title("Multi-Line")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=0) # loc (0-10)  shows legend at particular place
plt.legend(loc="center left") # also like this
plt.show()
```

## Customizing Graph

```python
# Customizing:-
# Line, Color, Marker
```

### Color

```python
plt.plot(x,[x**2 for x in x],"r", label = r'$y = x^2$')
plt.plot(x,[x**3 for x in x],"y", label = r'$y = x^3$')
plt.plot(x,[x**4 for x in x],"m", label = r'$y = x^4$') 

plt.title("Multi-Line")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=0)
plt.show()
```

### Line Style

```python
plt.plot(x,[x**2 for x in x], "r--", label=r'$y = x^2$')
plt.plot(x,[x**3 for x in x], "y-", label=r'$y = x^3$')
plt.plot(x,[x**4 for x in x], "m:", label=r'$y = x^4$') 

plt.title("Multi-Line")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=0)
plt.show()
```

### Marker Style

```python
x = np.arange(0.0,1.5,0.03)
plt.figure(figsize=(15,4))
plt.plot(x,[x**2 for x in x], "-s", label=r'$y = x^2$')
plt.plot(x,[x**3 for x in x], "-d", label=r'$y = x^3$')
plt.plot(x,[x**4 for x in x], "-+", label=r'$y = x^4$') 

plt.title("Multi-Line")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=0)
plt.show()
```

### Histogram

```python
# Histogram- Shows normal Distribution, hence used for scientific purposes
x = np.random.randn(1000)
plt.hist(x, edgecolor='white')
plt.show()
```

### Bar

```python
# Barplot - If categorical values
plt.bar([1,2,3],[3,2,5], color = ["tab:blue", "tab:red", "tab:green"])
plt.show()
```

### Pie Chart

```python
names = ["RS","VK","C","K","D"]
score = [45,35,25,40,35]
plt.pie(score,labels=names, autopct="%1.1f%%")
plt.show()
```

### Scatter

```python
# When two continuous column
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
```

### Add text at (x,y) position

```python
x = np.arange(1,10)
plt.plot(x)
plt.text(1,5,"LineGraph")
```

### Canvas Size

```python
plt.figure(figsize=(10,10))
names = ["RS","VK","C","K","D"]
score = [45,35,25,40,35]
plt.pie(score,labels=names, autopct="%1.1f%%")
plt.show()
```

### Subplots

```python
fig,axes = plt.subplots(2,2,figsize =(8,8))
axes[0,0].plot(x,[x*2 for x in x])
```
