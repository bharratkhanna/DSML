# Circle Area

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 100000000
```

```python
X = np.random.uniform(0,1,n)
Y = np.random.uniform(0,1,n)
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x=X,y=Y,c="tab:blue")
```

```python
circleX = list()
circleY = list()
r = 0.5

for i in range(n):
    distX = (X[i]-r)**2
    distY = (Y[i]-r)**2
    distance = (distX+distY)**0.5
    if distance < 0.5:
        circleX.append(X[i])
        circleY.append(Y[i])
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x=X,y=Y,c="tab:blue")
sns.scatterplot(x=circleX,y=circleY,c="tab:orange")

#sns.scatterplot(x=circleX,y=circleY, palette="orange")
```

```python
circle_pts = len(circleX) + len(circleY)
total_pts = len(X) + len(Y)

"""
circle area = pi*r**2 
square area = (2*r)**2 
circle/square = pi/4
4 * circle/square = pi
"""

pi = 4 * circle_pts/total_pts
pi,22/7
```
