**Q)** If a robber is travelling at speed 180 km/hr and sherrif car is travelling at the speed of 15 km/hr. Then at what distance the sheriff will catch the robber if sherrif's car started moving 5 minutes after robber's car?

**Solution:-**

$$150 \text{ km/hr} = \frac{150}{60} \text{ km/min} = 2.5 \text{ km/min}$$

$$180 \text{ km/hr} = \frac{180}{60} \text{ km/min} = 3 \text{ km/min}$$

Distance travelled by robber $d = 2.5t$

Distance travelled by sherrif $d = 3(t-5)$

$$2.5t = 3t -15 $$

$$15 = 0.5t $$

$$\boldsymbol {30 = t}$$

$$d = 30 * 2.5$$

$$ \boldsymbol{d = 75}$$

_So sherrif catch up with robber at distance 75 Km in 30 mins_

#### Plotting above equations on a graph

```python
t = np.linspace(0,40,1000) # Plots value for t in a uniform increements, so it generates 1000 numbers between 0 & 40

# Distance travelled by robber
d_r = 2.5 * t

# Distance travelleed by Sherrif
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title("Police Chase")
plt.xlabel("Time(minutes)")
plt.ylabel("Distance(Km)")

ax.set_xlim([0,40])
ax.set_ylim([0,100])

ax.plot(t,d_r, c="tab:orange")
ax.plot(t,d_s, c="tab:blue")
ax.scatter(30,75,c="tab:green")

plt.axvline(x=30,color="tab:purple",linestyle="--")
plt.axhline(y=75,color="tab:purple",linestyle="--")
plt.legend(["Robber","Sherrif"])
plt.show()
```

### Norms

They are functions that quantify vector magnitude(length of the given vector).

#### $\boldsymbol{L^2 Norm:}$

### $$\| x \|_2 = \sqrt{\sum_i{{x_i}^2}}$$

```python
x = [25,2,3]
norm = (25**2 + 2**2 + 3**2) ** 0.5
norm
```

```python
# Pytorch

x_pt = torch.tensor([25,2,3],dtype=torch.float16)
torch.norm(x_pt)
```

```python
# Tensorflow

x_tf = tf.Variable([25,2,3],dtype=tf.float16)
tf.norm(x_tf)
```

#### Unit Vectors

- Special case of vector where length is 1
- So if $\|x\|=1, x$ is a unit vector

#### $\boldsymbol{L^1 Norm:}$

### $$\| x \|_1 = {\sum_i{{x_i}}}$$

- Varies linearly at all locations whether near or far from origin.
- Can help distinguishing between zero and non zero elements

```python
x = [25,3,5]

l1_norm = 25 + 3 +5
l1_norm
```

#### $\boldsymbol{L_2^2 Norm:}$

### $$\| x \|_2^2 = {\sum_i{{x_i}^2}}$$

- Computationally cheaper to use than $L^2$ Norm
    - Squared $L^2$ norm equals to simply $x^Tx$
    - Derrivative (used to train many algorithms) of element x requires that element alsone, wheras L^2 norm requires x vector.
- Downside, grows slowly near origin, xo can't be used if distingushing between zero and non zero is important.

```python
x = [25,3,5]

squared_norm = (25**2 + 3**2 + 5**2)
squared_norm
```

#### Max Norm $\boldsymbol L^\infty$ :

### $$\| x \|_\infty = \max_{i} |x_i|$$

- Returns the largest magnitude element

```python
x = [25,3,5]

max(x)
```

#### Basis Vectors

- Can be scaled to represent any vector in a given vector space
- Typically use unit vectors along axes of vector space
- $i(1,0), j(0,1)$, hence new vector $v$ can be sum of both _(without effecting each other opposite axis)_.

### $$v = 1.5i + 2j$$

#### Orthoganal vectors

- $x$ and $y$ are orthogonal vectors if $x^Ty = 0$
- This means they are $90^\circ$ angle to each other (assuming non zero norms)
- $n$ dimensional space has max $n$ mutually orthoganal vectors
- **Orthonormal** vectors are orthoganal and all have unit norm
    - Basis Vectors are an example.

```python
# Pytorch
i_pt = torch.tensor([[1,0]])
j_pt = torch.tensor([[0,1]])
i_pt,j_pt
```

```python
torch.mul(i_pt,j_pt)
```

```python
# Tensorflow
i_tf = tf.Variable([[1,0]])
j_tf = tf.Variable([[0,1]])
i_tf, j_tf
```

```python
tf.multiply(i_tf,j_tf)
```

- Individual scalar elelments denoted in uppercase italics only
